3. Teniendo como base el siguiente codigo:
* Muestra el nombre del primer producto.
* Agrega un nuevo producto al inventario.
* Disminuye el stock de “Leche” en 2 unidades (venta).
* Calcula el valor total del inventario (precio * stock por producto).
* Muestra los productos cuyo stock sea menor a 10.
* Aumenta el precio de todos los productos en un 10%.
* Redondea los precios con round(p["precio"], 2) para evitar decimales largos.

* Elimina un producto del inventario si su stock es 0.
* Ordena la lista por precio ascendente.


```python
# Lista de productos en una tienda
inventario = [
    {"nombre": "Pan", "precio": 1500, "stock": 30},
    {"nombre": "Leche", "precio": 3500, "stock": 15},
    {"nombre": "Café", "precio": 8000, "stock": 10},
]

# Actualización del estado
#Agrega un nuevo producto al inventario.
inventario.append({"nombre": "Agua", "precio": 1845.34, "stock": 5})

#Disminuye el stock de “Leche” en 2 unidades (venta).
inventario[1]["stock"] -= 2

#Calcula el valor total del inventario (precio * stock por producto).
datos = []
for i in inventario:
  datos.append(i["precio"] * i["stock"])
suma = sum(datos)

#Aumenta el precio de todos los productos en un 10%.
for x in range(len(inventario)):
  inventario[x]["precio"] = inventario[x]["precio"] + (inventario[x]["precio"]* 0.1)

#Redondea los precios con round(p["precio"], 2) para evitar decimales largos.
for p in range(len(inventario)):
  inventario[p]["precio"] = round(inventario[p]["precio"], 2)
producto = input("¿Cuál producto necesita?: ")

#Elimina un producto del inventario si su stock es 0.
for k in range(len(inventario)):
  if inventario[k]["stock"] == 0:
    inventario.remove(inventario[k])

#Muestra los productos cuyo stock sea menor a 10.
for j in inventario:
  if j["stock"] < 10:
    print(j)

#Busca un producto por nombre (ingresado por el usuario) y muestra el producto encontrado, ten en cuenta que no puede estar.
encuentro = False
for k in range(len(inventario)):
  if inventario[k]["nombre"] == producto:
    encuentro = True
if encuentro == False:
  print("No se encuentra")
else:
  print("Si se encuentra en el inventario")

#Ordena la lista por precio ascendente(usando insertion sort).
for i in range(1, len(inventario)):
    actual = inventario[i]
    j = i - 1
    while j >= 0 and inventario[j]["precio"] > actual["precio"]:
        inventario[j + 1] = inventario[j]
        j -= 1
    inventario[j + 1] = actual

# Operaciones básicas

#Muestra el nombre del primer producto.
print(inventario[0]["nombre"])

print(inventario)

print(suma)

print("Productos en orden ascendente: ", inventario)
```

* Calcula la distancia total recorrida dadas las posiciones anteriores


```python
import math
posiciones = [(0, 0), (1, 2), (3, 4)]
distancia_total = 0

for i in range(len(posiciones) - 1):
    x1, y1 = posiciones[i]
    x2, y2 = posiciones[i + 1]

    distancia = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)


    distancia_total += distancia

print("Distancia total recorrida:", distancia_total)
```

* Muestra cuántos libros están prestados y cuántos disponibles.
* Agrega un nuevo libro.
* Marca como prestado el libro “1984”.
* Muestra el estado actualizado del sistema (usa print(biblioteca)).


```python
biblioteca = {
    "libros": [
        {"titulo": "1984", "autor": "George Orwell", "prestado": False},
        {"titulo": "El Principito", "autor": "Antoine de Saint-Exupéry", "prestado": True},
    ],
    "usuarios": ["Ana", "Luis", "María"]
}

#Muestra cuántos libros están prestados y cuántos disponibles.
disponibles = 0
for i in biblioteca["libros"]:
  if i["prestado"] == True:
    disponibles += 1
print("hay",disponibles,"libros disponibles")

#Agrega un nuevo libro.
biblioteca["libros"].append({"titulo": "Romeo y Julieta", "autor": "William Shakespeare", "prestado": True})
print(biblioteca["libros"])

#Marca como prestado el libro “1984”.
for k in range(len(biblioteca['libros'])):
  if biblioteca["libros"][k]["titulo"] == "1984":
    biblioteca["libros"][k]["prestado"] = True
print(biblioteca)
```
