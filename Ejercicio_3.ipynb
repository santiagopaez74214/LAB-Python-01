1. Teniendo en cuenta el siguiente codigo
* Filtrar solo los usuarios cuyo correo termine en .biz.
* Mostrar el nombre de estos usuarios


```python
import requests

# Llamado a una API p√∫blica
url = "https://jsonplaceholder.typicode.com/users"
response = requests.get(url)
usuarios = response.json()
print(usuarios)
# Procesamiento imperativo
nombres = []
for u in usuarios:
    nombres.append(u["name"])

print("Usuarios encontrados:")
for n in nombres:
    print("-", n)

#Filtrar solo los usuarios cuyo correo termine en .biz.
nombres_filtrados = []
for u in usuarios:
  if u["email"].endswith(".biz"):
    nombres_filtrados.append(u["name"])
print("Usuarios encontrados:")

#Mostrar el nombre de estos usuarios
for n in nombres_filtrados:
    print("-", n)
```

2. Teniendo en cuenta el siguiente codigo:

* Agregar condicionales adicionales (if / elif / else): Si el Pok√©mon tiene m√°s de un tipo, mostrar "Pok√©mon de tipo mixto", si el peso es mayor a 100, mostrar "Pok√©mon de gran tama√±o", En caso contrario, "Pok√©mon ligero".
* Usar for para listar las habilidades una por una
* A√±ade algun otro uso que consideres interesante para la respuesta de la api


```python
import requests

# Entrada de datos
nombre_pokemon = input("Ingrese el nombre de un Pok√©mon: ").lower()
tipo_pokemon = input("Ingrese el tipo de Pok√©mon a buscar: ").lower()

# Llamado a la API
url = f"https://pokeapi.co/api/v2/pokemon/{nombre_pokemon}"
respuesta = requests.get(url)

# Verificaci√≥n del estado de la respuesta
if respuesta.status_code == 200:
    datos = respuesta.json()

    # Modelado del estado del Pok√©mon
    pokemon = {
        "nombre": datos["name"],
        "altura": datos["height"],
        "peso": datos["weight"],
        "tipos": [t["type"]["name"] for t in datos["types"]],
        "habilidades": [h["ability"]["name"] for h in datos["abilities"]]
    }

    # Salida del estado actual
    print(url)
    print("\nüìò Informaci√≥n del Pok√©mon")
    print(f"Nombre: {pokemon['nombre'].capitalize()}")
    print(f"Altura: {pokemon['altura']}")
    print(f"Peso: {pokemon['peso']}")
    print(f"Tipos: {', '.join(pokemon['tipos'])}")
    print(f"Habilidades: {', '.join(pokemon['habilidades'])}")

    print(f"Tipos: {', '.join(pokemon['tipos'])}")
    #Agregar condicionales adicionales (if / elif / else): Si el Pok√©mon tiene m√°s de un tipo, mostrar "Pok√©mon de tipo mixto", si el peso es mayor a 100, mostrar "Pok√©mon de gran tama√±o", En caso contrario, "Pok√©mon ligero".
    if len(pokemon['tipos']) > 1:
        print("Pok√©mon de tipo mixto")
    else:
        print("Pok√©mon de tipo √∫nico")

    if pokemon['peso'] > 100:
        print("Pok√©mon de gran tama√±o")
    else:
        print("Pok√©mon ligero")

    #Usar for para listar las habilidades una por una
    print("\nHabilidades:")
    habilidades_lista = pokemon['habilidades']
    for i in range(len(habilidades_lista)):
        print(f"  {i+1}. {habilidades_lista[i].replace('-', ' ').title()}")

    #Otro uso - buscar los pok√©mones que tengan un tipo en espec√≠fico
    print(f"\nBuscando Pok√©mon del tipo: {tipo_pokemon.upper()}")

    url_tipo = f"https://pokeapi.co/api/v2/type/{tipo_pokemon}"
    respuesta_tipo = requests.get(url_tipo)

    if respuesta_tipo.status_code == 200:
        datos_tipo = respuesta_tipo.json()
        print(f"\nTODOS los Pok√©mon de tipo {tipo_pokemon.upper()}:")

        pokemones = datos_tipo['pokemon']
        for i in range(len(pokemones)):
            nombre_pokemon = pokemones[i]['pokemon']['name']
            print(f"  {i+1}. {nombre_pokemon.replace('-', ' ').title()}")

        total = len(pokemones)
        print(f"\nESTAD√çSTICAS:")
        print(f"Total de Pok√©mon de tipo {tipo_pokemon}: {total}")

    else:
        print(f"Tipo '{tipo_pokemon}' no encontrado.")
else:
    print("Pok√©mon no encontrado. Verifique el nombre ingresado.")
```
