2. Teniendo en cuenta el siguiente codigo:
* Calcula la suma de los primeros n números naturales.
* Imprime solo los números pares usando if.
* Agrega un acumulador y muestra el resultado final.
* Usando el ejercicio anterior de las temperaturas, encuentra las temperaturas mayores al promedio


```python
suma = 0
#Calcula la suma de los primeros n números naturales.
for i in range(1, 6):
    suma += i
    print(f"Iteración número {i}",end=",")
print(f"\n{suma}")

#Imprime solo los números pares usando if.
suma = 0
suma_pares = 0
for i in range(1, 6):
    suma += i
    if i % 2 == 0:
      suma_pares += i
      print(f"Iteración número {i}",end=",")
print(f"\n{suma}")
print(f"{suma_pares}")

#Usando el ejercicio anterior de las temperaturas, encuentra las temperaturas mayores al promedio
temperaturas = [22.5, 23.0, 21.8, 24.3, 25.1]
promedio = sum(temperaturas) / len(temperaturas)
print("Mayores al promedio")
for i in temperaturas:
  if i > promedio:
    print(i)
```

4. Teniendo en cuenta el siguiente codigo:
* Agrega otra variable para acumular los impares.
* Representa el estado final como un diccionario (suma_pares y suma_impares)
* Explica cómo el estado cambia dentro del bucle.


```python
suma_pares = 0
#Agrega otra variable para acumular los impares.
suma_impares = 0
for i in range(1, 11):
    if i % 2 == 0:
      suma_pares += i
    else:
      suma_impares += i

print("Suma de los pares:", suma_pares)
print("Suma de los impares:", suma_impares)
#Representa el estado final como un diccionario (suma_pares y suma_impares)
diccionario = {"suma_pares": suma_pares, "suma_impares": suma_impares}
print(diccionario)
```

El estado cambia dentro del bucle que nuestra variable iteradora va aumentando de uno en uno y se va agregando a los acumuladores dadas las condiciones que cumple.

5. Teniendo en cuenta el siguiente codigo:
* Usa continue para saltar productos con precio negativo o cantidad menor a 0.
* Muestra el producto más caro usando max().
* Representa el estado final del inventario como lista de diccionarios.



```python
productos = []
while True:
    nombre = input("Ingrese producto (o 'fin' para salir): ")
    if nombre == "fin":
        break
    precio = float(input("Precio: "))
    #Usa continue para saltar productos con precio negativo o cantidad menor a 0.
    if precio < 0:
      continue
    cantidad = int(input("Cantidad: "))
    productos.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})

# Mostrar total
total = 0
for p in productos:
    total += p["precio"] * p["cantidad"]

#Muestra el producto más caro usando max().
max_precio = 0
for p in productos:
    if p["precio"] > max_precio:
        max_precio = p["precio"]
print("Precio máximo:", max_precio)


print("Total del inventario:", total)

#Representa el estado final del inventario como lista de diccionarios.
print("\nInventario de productos:")
for p in productos:
  print(f"- {p['nombre']}, Precio: ${p['precio']}, Cantidad: {p['cantidad']}")

```
